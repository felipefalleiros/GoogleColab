```python
from bs4 import BeautifulSoup

import requests

url = "https://www.gov.br/esocial/pt-br/documentacao-tecnica/manuais/leiautes-esocial-v-1-1-beta/tabelas.html#25" # Alterar para a URL desejada
html = requests.get(url).content
soup = BeautifulSoup(html, 'html.parser')
```

```python
tabela = soup.find('table', id=25) # Alterar para o ID da tabela desejada
tabela
```

```python
data = []
headers = [th.text.strip() for th in tabela.find_all('th')] # Cabeçalhos da tabela
data.append(headers) # Adiciona os cabeçalhos como a primeira linha
for row in tabela.find_all('tr'):
    cells = row.find_all('td') # Obtém todas as células
    if cells: # Verifica se a linha tem células
        row_data = [cell.text.strip() for cell in cells]
        data.append(row_data)
```

```python
    df = pd.DataFrame(data)
    df.to_excel("tabela_extraida.xlsx", index=False, header=False) # index=False para não incluir o índice, header=False para não incluir o cabeçalho padrão
    print("Tabela salva em tabela_extraida.xlsx")
```
